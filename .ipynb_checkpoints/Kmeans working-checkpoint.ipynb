```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

file_name = './motor_collisions.csv'
xcol = ''
ycol = ''
xcolt = xcol + 't'
ycolt = ycol + 't'


# this is my helper function to help deal with the times
def hr_min_to_dec(st):
    hr = int(st[:st.find(':')])
    mi = int(st[st.find(':') + 1:])/60
    return str(hr + mi)

data = pd.read_csv(file_name, usecols=['CRASH TIME', 'NUMBER OF PERSONS KILLED', 'NUMBER OF PERSONS INJURED', 'NUMBER OF MOTORIST INJURED']);

# I also only want to use data for the number of people injured or number of people killed, and will cluster on those
data = data.transform({'CRASH TIME': hr_min_to_dec , 'NUMBER OF PERSONS KILLED': lambda x : x, 'NUMBER OF PERSONS INJURED': lambda x : x, 'NUMBER OF MOTORIST INJURED': lambda x : x}, axis=0)

# the last step made all of the crash times stringy, so lets undo that now
data['CRASH TIME'] = data['CRASH TIME'].map(float)

data
```

```python
# remove any NaN's
data.dropna(inplace=True)
data
```

```python
data.describe()
```

```python
# scale our inputs
scaler = StandardScaler()
data[['crashTime', 'numKilled', 'numInjured', 'numMotorInjured']] = scaler.fit_transform(data[['CRASH TIME', 'NUMBER OF PERSONS KILLED', 'NUMBER OF PERSONS INJURED', 'NUMBER OF MOTORIST INJURED']])
data
```

```python
# create number of clusters

def optimize_k_means(data, max_k):
    means = []
    inertias = []
    
    for k in range(1, max_k):
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(data)
        
        means.append(k)
        inertias.append(kmeans.inertia_)
    fig = plt.subplots(figsize=(10,5))
    plt.plot(means, inertias, 'o-')
    plt.xlabel('Num clusters')
    plt.ylabel('Inertia')
    plt.grid(True)
    plt.show()
```

```python
optimize_k_means(data[['crashTime', 'numMotorInjured']], 6)
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data[['crashTime', 'numMotorInjured']])
data['3k'] = kmeans.labels_
```

```python
data
```

```python
plt.scatter(x=data['CRASH TIME'], y=data['NUMBER OF MOTORIST INJURED'], c=data['3k'])
plt.xlabel('Crash Times')
plt.ylabel('Number of Motorists Injured')
plt.show()
```
